```python
#Importando a função docx2python para abrir arquivos DOCX
from docx2python import docx2python

lista_tabelas = [] #Lista para armazenar o conteúdo recuperado

#Para cada arquivo lido armazenar apenas a tabela após a segunda linha na lista criada
for y in lista_nomes:
    conteudos = docx2python(caminho_arquivos + '//' + y)
    lista_tabelas += conteudos.body[1][2:]

lista_tabelas
```

```python
import pandas as pd

tabela = pd.DataFrame(lista_tabelas, columns = ['Dia avaliado','Filial 1', 'Filial 2', 'Filial 3', 'Matriz'])
tabela
```

```python
#Importando o pandas
import pandas as pd

#Lista com os nomes das colunas para o dataset
colunas_novas = ['Dia avaliado','Filial 1', 'Filial 2', 'Filial 3', 'Matriz']

#Transformando a lista em dataframe do pandas 
tabela = pd.DataFrame(lista_tabelas, columns = colunas_novas).applymap(lambda x : x[0])
tabela
```
